First, our import statements.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mping
import numpy as np
from PIL import Image, ImageOps
```

Next, lets read in a `.jpg` picture that we want to investigate.

```python
image = mping.imread("pattern.jpg")
imgplot = plt.imshow(image)
plt.show()
```

If you look closely, you can see a white grid pattern. Let's try to remove this. First, let's put it to a greyscale to we dont have to worry about the color data.

```python
img = np.array(ImageOps.grayscale(Image.open('pattern.jpg')))
plt.imshow(img, "gray")
```

Next, we can take the 2D FFT of the grid and center it.

```python
fimg = np.fft.fft2(img)
fimgcenter = np.fft.fftshift(fimg)
plt.imshow(np.log(1+np.abs(fimgcenter)), "gray")
```
